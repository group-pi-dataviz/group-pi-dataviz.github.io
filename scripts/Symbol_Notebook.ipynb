```python
import pandas as pd
```

```python
data = pd.read_csv('apac_data.csv', sep=';')
data.head()
```

```python
af_bat = data[(data['COUNTRY'] == 'Afghanistan') & (data['SUB_EVENT_TYPE'] == 'Armed clash')]
af_bat.head()
```

```python
used_data = af_bat[['WEEK', 'ADMIN1', 'FATALITIES', 'CENTROID_LATITUDE', 'CENTROID_LONGITUDE']]
def month_to_eng(month):
    match month:
        case 'gennaio': return 'January'
        case 'febbraio': return 'February'
        case 'marzo': return 'March'
        case 'aprile': return 'April'
        case 'maggio': return 'May'
        case 'giugno': return 'June'
        case 'luglio': return 'July'
        case 'agosto': return 'August'
        case 'settembre': return 'September'
        case 'ottobre': return 'October'
        case 'novembre': return 'November'
        case 'dicembre': return 'December'

def month_to_num(month):
    match month:
        case 'January': return 1
        case 'February': return 2
        case 'March': return 3
        case 'April': return 4
        case 'May': return 5
        case 'June': return 6
        case 'July': return 7
        case 'August': return 8
        case 'September': return 9
        case 'October': return 10
        case 'November': return 11
        case 'December': return 12

used_data['YEAR'] = used_data['WEEK'].apply(lambda x: int(x.split('-')[2]))
used_data['MONTH'] = used_data['WEEK'].apply(lambda x: month_to_num(month_to_eng(x.split('-')[1])))
used_data['DAY'] = used_data['WEEK'].apply(lambda x: int(x.split('-')[0]))

print(len(used_data))

used_data.head()
```

```python
used_data.pivot_table(['ADMIN1'])
```

```python
used_data.to_csv('af_geobat.csv', sep=';')
```
